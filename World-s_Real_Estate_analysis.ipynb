```python
#Import the libraries to be used

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the data from csv file
world_real_estate_data=pd.read_csv("world_real_estate_data.csv")
```

```python
#Size of the data
world_real_estate_data.shape
```

```python
#Show the data
world_real_estate_data.head(20)
```

# Data Cleaning

```python
world_real_estate_data.duplicated().sum()
```

```python
world_real_estate_data.drop_duplicates()
```

```python
#Drop unnecessary columns and the columns that contained more than 80% null values
world_real_estate_data=world_real_estate_data.drop(columns=["image","url","apartment_living_area"])
```

```python
world_real_estate_data.info()
```

```python
#Percentage of the nulls for each column
null_percentage=world_real_estate_data.isnull().sum()/len(world_real_estate_data)*100
null_percentage
```

```python
world_real_estate_data.dtypes
```

```python
#There is a few missing values in ["country"] column, will fill it with the country that provided in the ["title"] column
world_real_estate_data["country"] = world_real_estate_data["title"].apply(lambda value: value.split(",")[1])
```

```python
world_real_estate_data.info()
```

```python
# Drop rows containing null values in ["apartment_total_area"], ["price_in_USD"] and ["location"] because the null values contained in these two columns do not exceed 5%
world_real_estate_data = world_real_estate_data.dropna(subset=["apartment_total_area","price_in_USD","location"])
```

```python
#Fill the null values in ["apartment_bathrooms"] with 1
world_real_estate_data["apartment_bathrooms"] = world_real_estate_data["apartment_bathrooms"].fillna(1)
```

```python

#fill_bedrooms function
def fill_bedrooms(row):
    if pd.isnull(row['apartment_bedrooms']): #Check if apartment_bedrooms is null
        if row['apartment_rooms']==1: #If apartment_rooms is 1
            return 1 #Fill it with 1
        else:
            return row['apartment_rooms']-1  #Else fill with apartment_rooms-1
    # If apartment_bedrooms is not null
    return abs(row['apartment_bedrooms'])  #Convert any negative value to positive


world_real_estate_data['apartment_bedrooms'] = world_real_estate_data.apply(fill_bedrooms, axis=1)
```

```python
#Fill the null values in ["apartment_bedrooms"] with 1
world_real_estate_data["apartment_bedrooms"] = world_real_estate_data["apartment_bedrooms"].fillna(1)
```

```python
#Define the function to fill null values in ['apartment_rooms'] from ['apartment_bedrooms']+1
def fill_apartment_rooms(row):
    if pd.isna(row['apartment_rooms']):  #Check if apartment_rooms is null
        return row['apartment_bedrooms']+1  #Fill with apartment_bedrooms+1
    return row['apartment_rooms']  #Return the original value if apartment_rooms is not null

world_real_estate_data['apartment_rooms'] = world_real_estate_data.apply(fill_apartment_rooms, axis=1)
```

```python
world_real_estate_data["apartment_floor"] = world_real_estate_data["apartment_floor"].fillna(0)
```

```python
world_real_estate_data["building_total_floors"] = world_real_estate_data["building_total_floors"].fillna(1)
```

```python
null_percentage=world_real_estate_data.isnull().sum()/len(world_real_estate_data)*100
null_percentage
```

```python
world_real_estate_data.shape
```

```python
def fill_invalid_year(row, median_year):
    # Check if ['building_construction_year'] is greater than 2025 or is null
    if row['building_construction_year']>2025 or pd.isna(row['building_construction_year']):
        return median_year #Replace with median
    return row['building_construction_year']  #Keep the original value otherwise

#Calculate the median of ['building_construction_year'] 
median_year=world_real_estate_data['building_construction_year'].median()


world_real_estate_data['building_construction_year']=world_real_estate_data.apply(fill_invalid_year, axis=1, median_year=median_year)
```

```python
world_real_estate_data.isnull().sum()
```

```python
world_real_estate_data['apartment_total_area'] = world_real_estate_data['apartment_total_area'].dropna().str.replace('m²',"")
```

```python
world_real_estate_data
```

```python
#Function to clean ["apartment_total_area"] values
def clean_area(value):
    #Remove 'm²' and any spaces
    cleaned_value = ''.join(value.replace('m²', '').split())
    #Convert the cleaned value to a float
    return float(cleaned_value)

world_real_estate_data['apartment_total_area'] = world_real_estate_data['apartment_total_area'].apply(clean_area)
```

```python
world_real_estate_data.isnull().sum()
```

```python
world_real_estate_data=world_real_estate_data.drop(columns=["title"])
```

# Data Analysis

```python
world_real_estate_data["country"].unique()
```

```python
country_counts = world_real_estate_data['country'].value_counts()
sns.set_theme()
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Country')
plt.ylabel('Count of Apartments')
plt.title('Count of Apartments by Country')

plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=world_real_estate_data, x='country', y='price_in_USD')
plt.title('Price in USD by Country', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Price in USD', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
start_year=2010
end_year=2025
filtered_data=world_real_estate_data[(world_real_estate_data['building_construction_year']>=start_year)&(world_real_estate_data['building_construction_year']<=end_year)]
if filtered_data.empty:
    print("Nothing")
else:
    plt.figure(figsize=(10,6))
    sns.lineplot(data=filtered_data, x='building_construction_year', y='price_in_USD')
    plt.title(f'Building Construction Year vs Price in USD ({start_year} to {end_year})',fontsize=16)
    plt.xlabel('Building Construction Year', fontsize=12)
    plt.ylabel('Price in USD', fontsize=12)

   
    plt.show()
```

```python
filtered_data = world_real_estate_data[(world_real_estate_data['apartment_total_area']>=25)&(world_real_estate_data['apartment_total_area']<=500)]
sns.set_theme()
plt.figure(figsize=(10, 6))
plt.plot(filtered_data['apartment_total_area'],filtered_data['price_in_USD'],'x',label='Price vs Area')
plt.title('Price in USD vs Apartment Total Area (25 - 500 sqm)', fontsize=16)
plt.xlabel('Apartment Total Area (sqm)', fontsize=12)
plt.ylabel('Price in USD', fontsize=12)

plt.tight_layout()
plt.show()
```
